```python
d = {'user_id': [1, 2, 5, 14, 15],
             'address': ["Acassuso", "Martinez", "Tigre", "san Isidro", "Vicente Lopez"],
            'country': ["AR", "AR", "AR", "AR", "AR"]}
```

```python
addresses = pd.DataFrame(data=d)
addresses
```

```python
# Inner Join
pd.merge(user, addresses, left_on=['id'], right_on=['user_id'])
```

```python
# Left Join
pd.merge(user, addresses, how="left", left_on=['id'], right_on=['user_id'])
```

```python
# Right Join
pd.merge(user, addresses, how="right", left_on=['id'], right_on=['user_id'])
```

```python
# Full Join
pd.merge(user, addresses, how="outer", left_on=['id'], right_on=['user_id'])
```

```python
# Producto cartesiano, todas las combinaciones posibles 
pd.merge(user, addresses, how="cross")
```

# Importar de set de datos

Para hacer la importacio vamos a usar https://www.kaggle.com/ <br />
vamos a usar el del dolar: https://www.kaggle.com/datasets/guidoprd/dolar-oficial-arg-xx082011-xx082021

```python
dolar = pd.read_csv("dataset/USD-ARG 08-2011_08-2021.csv")
dolar
```

```python
dolar['Fecha'] = pd.to_datetime(dolar['Fecha'], format='%d.%m.%Y')
```

```python
dolar
```

```python
# (dolar['Fecha'] - dolar['Fecha'][len(dolar) - 1])  me lo devuelve en format datatime, pero lo necesito numerico
dolar["days"] = (dolar['Fecha'] - dolar['Fecha'][len(dolar) - 1]).dt.days
```

```python
dolar
```

```python
dolar["cash"] = dolar["Último"].str.replace(',','.').astype(float)
```

```python
dolar
```

```python
plt.plot(dolar['days'], dolar['cash'])
plt.show()
```
